```python
import matplotlib.pyplot as plt

# Seleccionamos 5 columnas al azar para visualizar
random_columns = data.sample(n=5, axis=1)

# Dibujamos histogramas para las columnas seleccionadas
plt.figure(figsize=(15,10))

for i, col in enumerate(random_columns.columns, 1):
    plt.subplot(2, 3, i)
    data[col].hist(bins=50, color='blue', alpha=0.7)
    plt.title(f'Histograma de {col}')
    plt.xlabel('Valor del píxel')
    plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
# Análisis de frecuencia para el píxel 'pixel774'
frequency_pixel774 = data['pixel774'].value_counts().sort_index()

frequency_pixel774
```

El valor 0 (negro) aparece 27,963 veces, lo que es abrumadoramente el valor más común.
Los valores más altos, que indicarían píxeles más blancos, aparecen con mucha menos frecuencia.
Hay varios valores que aparecen solo una vez en todo el conjunto de datos.
Este análisis sugiere que la mayoría de las imágenes tienen un fondo oscuro (negro) en la posición correspondiente a pixel774, pero hay algunas imágenes que tienen píxeles más claros en esa posición.

```python
# Seleccionamos un subconjunto de 10 columnas al azar para la matriz de dispersión
subset_10_data = data.sample(n=10, axis=1)

# Dibujamos la matriz de dispersión para el subconjunto seleccionado
sns.pairplot(subset_10_data, diag_kind="kde", markers='.')
plt.suptitle('Matriz de Dispersión para 10 píxeles aleatorios', y=1.02)
plt.show()
```

Las gráficas en la diagonal muestran la distribución de valores para cada píxel individualmente.

Las gráficas fuera de la diagonal comparan la relación entre dos píxeles diferentes.

Como se puede observar, la mayoría de los puntos en las gráficas están concentrados cerca del valor 0, lo que indica que la mayoría de los píxeles tienen un valor cercano al negro. Las gráficas de densidad en la diagonal también reflejan esta observación
